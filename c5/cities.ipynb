The list of Polish cities is available in the cities-in-poland.csv file. Using Power Query, import the data. Then, create the following charts:

the total number of inhabitants in each province
the area of the largest city in each province

```python
import pandas as pd

cities = pd.read_csv("cities-in-poland.csv", delimiter=';')
cities
```

```python
plota = cities.groupby(by='Province')['Population'].sum()
```

```python
import matplotlib.pyplot as plt
plota.plot(kind="bar")
```

```python
cities.groupby(by=['Province'])['Areakm2'].max()
largest_city_in_province = cities.loc[cities.groupby('Province')['Areakm2'].idxmax()][['Province', 'City', 'Areakm2']]
largest_city_in_province
```

```python
cities2  = cities.groupby(by=['Province'])['Areakm2'].idxmax()
cities_max = cities.loc[cities2][["City", "Areakm2"]]
cities_max

```

```python
cities2  = cities.groupby(by='Province')['Areakm2'].max()


cities2.plot(kind="pie",autopct='%1.0f%%' )
```
