```python
import pandas as pd
import matplotlib.pyplot as plt
```

Create a new notebook. Then, based on the european-countries.csv file, create:

a bar chart with the population of Central European countries (sort the data by population)
a pie chart presenting the percentage of the regions' area.

```python
countries_data = pd.read_csv("european-countries.csv", index_col='Name')
countries_data
countries_data = countries_data.query("Region == 'Central Europe'").sort_values(by="Population")
countries_data['Population'].plot(kind='bar', title="Population by country")
```

```python
countries_data = pd.read_csv("european-countries.csv", index_col='Name')
countries_data = countries_data.groupby("Region")["Area"].sum()
countries_data.plot(kind='pie', autopct='%1.0f%%', ylabel='', title="Area by Region")
```
