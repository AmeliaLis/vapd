```python
import pandas as pd
import matplotlib.pyplot as plt

employees_data = pd.read_csv("employees.csv")
employees_data
```

```python
test = employees_data.groupby("sex")["age"].mean()
test
```

```python
test.plot(title="Average age by sex", kind="bar", color='skyblue')
```

```python
test_2 = employees_data.groupby("sex")[['salary','sex']]
test_2.plot(kind="hist", layout=(2,1))
```

```python
#test_2 = employees_data.groupby("sex")[["salary","age"]]
test_2 = employees_data.query("sex=='Male'")[["salary"]]
test_2.plot(kind='hist', xlabel="Salary", title="Histogram for Male salary")
```

```python
test_2 = employees_data.query("sex=='Female'")[["salary"]]
test_2.plot(kind='hist', xlabel="Salary", title="Histogram for Female salary")
```
