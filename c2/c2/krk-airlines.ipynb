1. The files krk-airlines.csv, krk-flights.csv and krk-passengers.csv contain data about flights from Krakow Airport. In a separate notebook, calculate and display:

    * list including flight number and destination (two columns)
    * flight list with the full name of the airline and the name of the aircraft
    * a list of passengers on a flight to London sorted by surname
    * a list of women flying to Paris ordered by surname
    * number of passengers on the flight to Berlin
    * number of men flying from Krakow
    * number of passengers for each flight

```python
import pandas as pd
airlines = pd.read_csv('krk-airlines.csv')
flights = pd.read_csv('krk-flights.csv')
passengers = pd.read_csv('krk-passengers.csv')

airlines
```

```python
flight_airline = pd.merge(flights,airlines,on='airlineid')
flight_airline
```

```python
passengers_flights = pd.merge(passengers, flights, on='flight')
passengers_flights.query("to == 'London'").sort_values(by='surname')
```

```python
passengers_flights = pd.merge(passengers, flights, on='flight')
passengers_flights.query("to == 'Paris' and gender == 'Female'").sort_values(by='surname')
```

```python
passengers_flights = pd.merge(passengers, flights, on='flight')
passengers_flights.query("to == 'Berlin'").loc[:,['to','surname']].groupby(['to']).count()
passengers_flights.query("to == 'Berlin'")['name'].count()
```

```python
passengers_flights.query("gender == 'Male' and start == 'Krakow'")['flight'].count()
```

```python
passengers_flights.groupby('flight')['name'].count()
passengers_flights.groupby('flight').size()
```
